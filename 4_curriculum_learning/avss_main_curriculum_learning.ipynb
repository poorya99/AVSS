```python
train =  [
    {'id': '1S-VLcoCQOXTw1AAyh1edeM5RINVtW-dI', 'output': 'tens_train_5_1.zip'}, {'id': '1blZ6nkVSQMd9_Bs3FZ_DwcBWu947ExUr', 'output': 'tens_train_5_2.zip'},
          {'id': '1moTzWYYOMfuM8w1MkPrRUZYPheY_b5mm', 'output': 'tens_train_5_3.zip'},
          {'id': '1y_y1fxFN0CGByhaSmF3LwXkO4A90SmKF', 'output': 'tens_train_4_1.zip'}, {'id': '1a-0oI_fHJZr-HRlJod5F0TVk6TeGj26d', 'output': 'tens_train_4_3.zip'}, {'id': '108gCW69kNWebE6F2GCFKWO-ZJUyS-i-i', 'output': 'tens_train_2_2.zip'}, {'id': '1Ny4vCJWIzUwC8DnQfH5p4B4KjaBoaEc1', 'output': 'tens_train_2_3.zip'}, {'id': '1fNUb7u_oK74pF583FxrPNKlNf17SyOYp', 'output': 'tens_train_3_1.zip'}, {'id': '1dv_P1xM1x4DruaqHQRXbTkzCqC03PxHA', 'output': 'tens_train_1_2.zip'}, {'id': '1yQHt3-UKICLfD3QUFUrOzD41safbbpDB', 'output': 'tens_train_1_1.zip'}, {'id': '1YlulmSQrvyesKiEkIxbp--4y41NUWdZL', 'output': 'tens_train_4_2.zip'}, {'id': '1jJhHUC8RoBfAy5b9WXEaEheg2KQRAdoZ', 'output': 'tens_train_2_1.zip'}, {'id': '1L8ZjazHvMlGUKrS1fmSi4yswh8SAcFsd', 'output': 'tens_train_3_3.zip'}, {'id': '1j0Tq-5Xpiks3Fvf10g1eG-P7m6lYnzE8', 'output': 'tens_train_3_2.zip'}, {'id': '1YDesdSlPoaUDk1ZgCx7RgLPsWPNPSavx', 'output': 'tens_train_1_3.zip'}
    ]
val =  [
    {'id': '1kOSO1WxN23EpCzvrwB3V5LbMK2zXkj_r', 'output': 'tens_val_1_1.zip'},
     {'id': '13DqHbsOa1hf1EHI1QOGJMvSr0ZYJzUQA', 'output': 'tens_val_2_1.zip'}, {'id': '1ucQ1MFHLBS1LNysKUNEzXh1v8pmtjeae', 'output': 'tens_val_1_3.zip'}, {'id': '1fLdbA4qrid44gnDY3W5JihzFv02TNPGI', 'output': 'tens_val_1_2.zip'}, {'id': '1Zfb8gj33x4GkkZS-am2sXSItZs3r0CRf', 'output': 'tens_val_2_2.zip'}, {'id': '18YkppMZBoPno1k_-cBO7N9IZ5LKgfXrx', 'output': 'tens_val_2_3.zip'}
    ]
test =  [
    {'id': '11M1d3Y-QKsRRtDeOtqsysi0pjxezvbhz', 'output': 'tens_test_1_1.zip'},
     {'id': '1p-MX2dUD5xqxERuouv_BW-rtsMV7-xze', 'output': 'tens_test_1_3.zip'}, {'id': '10oj7cnMkmHipAP2VGMKFIyRMItuoDQWd', 'output': 'tens_test_1_2.zip'}
    ]


import os
import io
import zipfile
from googleapiclient.discovery import build
from googleapiclient.http import MediaIoBaseDownload
from google.oauth2.service_account import Credentials


def download_and_extract_files(service_account_json, file_list, extraction_dir="."):
    """
    Downloads and extracts files based on a list of dictionaries containing file IDs and output names.

    Args:
        service_account_json (str): Path to the service account JSON key file for Google Drive API authentication.
        file_list (list of dict): List of dictionaries, each containing 'id' (file ID) and 'output' (output file name).
        extraction_dir (str): Directory to extract the files into. Defaults to the current directory.
    """
    # Authenticate using the service account JSON file
    credentials = Credentials.from_service_account_file(service_account_json, scopes=["https://www.googleapis.com/auth/drive"])
    service = build('drive', 'v3', credentials=credentials)

    os.makedirs(extraction_dir, exist_ok=True)

    for file in file_list:
        file_id = file["id"]
        output_path = os.path.join(extraction_dir, file["output"])

        # Download the file
        request = service.files().get_media(fileId=file_id)
        print(f"Downloading {file['output']}...")
        with io.FileIO(output_path, 'wb') as file_obj:
            downloader = MediaIoBaseDownload(file_obj, request)
            done = False
            while not done:
                status, done = downloader.next_chunk()
                print(f"Download {int(status.progress() * 100)}%.")

        # Check if the file is a ZIP file and extract it
        if output_path.endswith(".zip"):
            print(f"Extracting {output_path}...")
            with zipfile.ZipFile(output_path, 'r') as zip_ref:
                zip_ref.extractall(extraction_dir)

            # Remove the ZIP file after extraction
            print(f"Removing {output_path}...")
            os.remove(output_path)

    print(f"All files downloaded and extracted to {extraction_dir}.")

# Example usage
if __name__ == "__main__":

    service_account_json = "avss11.json"  # Path to your service account JSON key file
    extraction_dir = "."  # Directory to store extracted files
    download_and_extract_files(service_account_json, test, extraction_dir)
    download_and_extract_files(service_account_json, val, extraction_dir)
    # download_and_extract_files(service_account_json, train, extraction_dir)
```
